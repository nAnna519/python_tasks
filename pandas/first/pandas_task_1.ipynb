```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('adult.data.csv')
data.head()
```

**1. Сколько мужчин и женщин (признак *sex*) представлено в этом наборе данных?**

```python
data['sex'].value_counts()
```

**2. Каков средний возраст (признак *age*) женщин?**

```python
data[data['sex'] == 'Female']['age'].mean()
```

**3. Какова доля граждан Германии (признак *native-country*)?**

```python
data['native-country'].value_counts(normalize=True)['Germany']
```


**4-5. Каковы средние значения и среднеквадратичные отклонения возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год? **

```python
columns_to_show = ['age']

data.groupby(['salary'])[columns_to_show].agg([np.mean, np.std])
```


**6. Правда ли, что люди, которые получают больше 50k, имеют как минимум высшее образование? (признак *education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* или *Doctorate*)**

```python
pd.crosstab(data['education'], data['salary'])
```

**7. Выведите статистику возраста для каждой расы (признак *race*) и каждого пола. Используйте *groupby* и *describe*. Найдите таким образом максимальный возраст мужчин расы *Amer-Indian-Eskimo*.**

```python
columns_to_show = ['age']

data.groupby(['race'])[columns_to_show].describe()
```

```python
data.groupby(['sex'])[columns_to_show].describe()
```

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python

married_men = data['salary'][data['sex'] == 'Male'][(data['marital-status'] == 'Married-civ-spouse') | 
                             (data['marital-status'] == 'Married-AF-spouse') | 
                             (data['marital-status'] == 'Married-spouse-absent')].value_counts(normalize=True)['>50K']

print('Married men: ' + str(married_men))
alone_men = data['salary'][data['sex'] == 'Male'][(data['marital-status'] == 'Divorced') | 
                             (data['marital-status'] == 'Never-married') | 
                             (data['marital-status'] == 'Separated')| 
                             (data['marital-status'] == 'Widowed')].value_counts(normalize=True)['>50K']

print('Single men: ' + str(alone_men))
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
max_value = data['hours-per-week'].max()
print('Maximum: ' + str(max_value))

count_value = data[data['hours-per-week'] == max_value]['salary'].count()
print('People count: ' + str(count_value))

people_count_with_big_salary = data[data['hours-per-week'] == max_value][data['salary'] == '>50K']['age'].count()

percentages = people_count_with_big_salary / count_value
print('Percentage: ' + str(percentages))
```
